# Результаты статистического анализа

```python
def comparsion_plot(distr1_vs_distr2_results, 
                    marker_for_type, distr1_name, distr2_name, distr1_color, distr2_color):
    """
    Изображает результаты работы критерия отношения правдоподобия
    Строит график (loglikelihood ratio, p-value)
    
    :param distr1_vs_distr2_results: словарь с результатами критерия вида 
                                     {'name': [], 'R': [], 'pval': [], 'tectonic type': []}
    :param marker_for_type: словарь маркеров и цветов для разных тектонических типов
    :param distr1_name: название первого распределения
    :param distr1_name: название второго распределения
    :param distr1_color: цвет заливки для первого распределения 
    :param distr1_color: цвет заливки для второго распределения
    """
    n_samples = len(distr1_vs_distr2_results['R'])
    
    for i in range(n_samples):
        R = distr1_vs_distr2_results['R'][i]
        p = distr1_vs_distr2_results['pval'][i]
        label = distr1_vs_distr2_results['tectonic type'][i]

        plt.scatter(R, p,
                    marker=marker_for_type[label][0], 
                    color=marker_for_type[label][1],
                    edgecolors='black', 
                    linewidths=0.6)

    plt.xlabel('R, loglikelihood ratio')
    plt.ylabel('p-value')
    
    maximum = 1.05 * np.max( np.abs(distr1_vs_distr2_results['R']) ) 
    plt.xlim(-maximum, maximum)
    plt.ylim(-0.03, 1)
    plt.vlines(x=-0.05, ymin=-0.05, ymax=0.1, color='black', lw=2)
    plt.hlines(y=0.1, xmin=-maximum, xmax=maximum, color='black', lw=2)
    plt.fill_between([-maximum, maximum], y1=0.1, y2=1, alpha=0.4, color='C0')
    plt.fill_between([-maximum, 0], y1=-0.05, y2=0.1, alpha=0.4, color=distr2_color)
    plt.fill_between([0, maximum], y1=-0.05, y2=0.1, alpha=0.4, color=distr1_color)
    plt.title(distr2_name + '                  ' + distr1_name)
```

```python
powerlaw_vs_exponential_results = {'name': [], 'R': [], 'pval': [], 'tectonic type': []}

powerlaw_vs_lognormal_results = {'name': [], 'R': [], 'pval': [], 'tectonic type': []}

lognormal_vs_exponential_results = {'name': [], 'R': [], 'pval': [], 'tectonic type': []}
```

```python
for filename in tqdm(os.listdir(path_to_pickle_files)):
    file = open(path_to_pickle_files + filename, 'rb')
    data = pickle.load(file)
    file.close()

    t = data.stat_tests_lineaments_length  # таблица с результатами критерия   
    
    powerlaw_vs_exponential_results['name'].append(data.name)
    powerlaw_vs_exponential_results['R'].append(t['R']['powerlaw vs. exponential'])
    powerlaw_vs_exponential_results['pval'].append(t['p']['powerlaw vs. exponential'])
    powerlaw_vs_exponential_results['tectonic type'].append(data.tectonic_type)

    powerlaw_vs_lognormal_results['name'].append(data.name)
    powerlaw_vs_lognormal_results['R'].append(t['R']['powerlaw vs. lognormal'])
    powerlaw_vs_lognormal_results['pval'].append(t['p']['powerlaw vs. lognormal'])
    powerlaw_vs_lognormal_results['tectonic type'].append(data.tectonic_type)

    lognormal_vs_exponential_results['name'].append(data.name)
    lognormal_vs_exponential_results['R'].append(t['R']['lognormal vs. exponential'])
    lognormal_vs_exponential_results['pval'].append(t['p']['lognormal vs. exponential'])
    lognormal_vs_exponential_results['tectonic type'].append(data.tectonic_type)
```

```python
powerlaw_color = 'C8'
exp_color = 'C7'
lognormal_color = 'C1'

marker_for_type = {'Бластомилонит': ['o', 'C2'],  # тип маркера и цвет для разных тектонических типов
                   'Катаклазит': ['D', 'C3'],
                   'Бластокатаклазит': ['s', 'C4'],
                   'Вмещающая порода': ['X', 'C5'],
                   'Милонит': ['P', 'C6']}
```

```python
fig = plt.figure(figsize=(25, 8))

plt.subplot(1, 3, 1)
comparsion_plot(powerlaw_vs_exponential_results, 
                marker_for_type, 
                'Power-Law', 'Exponential', powerlaw_color, exp_color)

plt.subplot(1, 3, 2)
comparsion_plot(powerlaw_vs_lognormal_results, 
                marker_for_type, 
                'Power-Law', 'Log-Normal', powerlaw_color, lognormal_color)

plt.subplot(1, 3, 3)
comparsion_plot(lognormal_vs_exponential_results,  
                marker_for_type, 
                'Log-Normal', 'Exponential', lognormal_color, exp_color)
```

**Вывод:** 
распределение длин трещин обработанных образцов, в основном, подчиняется лог-нормальному закону

# Результаты топологического анализа

```python
import ternary
```

```python
n = len(os.listdir(path_to_pickle_files))  # число образцов
points = np.zeros((n, 3))
labels = np.zeros(n, dtype=object)

for i in tqdm(range(n)):
    filename = os.listdir(path_to_pickle_files)[i]
    file = open(path_to_pickle_files + filename, 'rb')
    data = pickle.load(file)
    file.close()
    
    points[i] = data.XIY_counts_normed
    labels[i] = data.tectonic_type
```

```python
# словарь: ключ - тектонический тип, 
# значение - спиcок кортежей, в каждом кортеже 3 элемента (относительное кол-во точек X, I, Y)
groups = {}

for i in range(n):
    label = labels[i]
    if label not in groups:
        groups[label] = [tuple(points[i])]
    else:
        groups[label].append(tuple(points[i]))
```

```python
rcParams['figure.figsize'] = 10, 10
figure, tax = ternary.figure(scale=1.0)
figure.set_facecolor('w')
tax.set_background_color('w')

for label in groups:
    tax.scatter(groups[label], 
                marker=marker_for_type[label][0], 
                color=marker_for_type[label][1],
                edgecolors='black', 
                linewidths=0.6,
                label=label)

tax.boundary(linewidth=1.0)
tax.ticks(axis='lbr', linewidth=1, multiple=0.1, tick_formats="%.1f")
tax.gridlines(multiple=0.2, color="black")
tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()
tax.legend(facecolor='white', fontsize=13.5, bbox_to_anchor=[0.072, 0.95]) 
tax.right_corner_label("X-Nodes", fontsize=18, offset=-0.00001)
tax.top_corner_label("                        I-Nodes", fontsize=20, offset=0.1)
tax.left_corner_label("           Y-Nodes", fontsize=20, offset=0.35)
tax.show()
```

**Вывод:**
область локализации образцов на тернарной диаграмме уширяется с увеличением степени преобразованности породы

# Связь топологии с энтропией

```python
entropy_values = np.zeros(n)
counts_normed_I = np.zeros(n)
```

```python
for i in tqdm(range(n)):
    filename = os.listdir(path_to_pickle_files)[i]
    file = open(path_to_pickle_files + filename, 'rb')
    data = pickle.load(file)
    file.close()
    
    entropy_values[i] = data.entropy_rho_az
    counts_normed_I[i] = data.XIY_counts_normed[1]
```

```python
fig = plt.figure(figsize=(8, 8))

plt.scatter(entropy_values, counts_normed_I, color='black')

grid = np.linspace(np.min(entropy_values) - 0.02 * np.min(entropy_values), 
                   np.max(entropy_values) + 0.02 * np.max(entropy_values), 100)
plt.xlabel('Энтропия')
plt.ylabel('Доля вершин типа I')
plt.xlim(np.min(grid), np.max(grid))
plt.ylim(0, 1)
plt.grid()
None
```

Корреляция топологии с энтропией

```python
rho, _ = sps.pearsonr(entropy_values, counts_normed_I)

print('Коэффициент корреляции:', rho)
```
