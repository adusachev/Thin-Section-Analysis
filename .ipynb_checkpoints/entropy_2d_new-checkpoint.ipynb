```python
def save_pickle(data, path_to_pickle_files):
    """
    Cохранение pickle объекта
    """
    name = data.name
    
    object_name = name + '_object'
    path_to_object = os.path.join(path_to_pickle_files, object_name) 
    file = open(path_to_object, 'wb')
    pickle.dump(data, file)
    file.close()
```

```python
def norm(array):
    """
    Нормирует количества точек X, I, Y от 0 до 1
    """
    n = len(array)
    array_normed = np.zeros(array.shape)
    for i in range(n):
        array_normed[i] = array[i, :] / np.sum(array[i, :])
        
    return array_normed
```

```python
counts = IXY[['X-nodes', 'I-nodes', 'Y-nodes']]
counts_normed = norm(counts.values)
counts_normed_I = counts_normed[:, 1]
```

```python
area1_names = IXY['name'].values[area1_indexes]
area2_names = IXY['name'].values[area2_indexes]
area3_names = IXY['name'].values[area3_indexes]
area4_names = IXY['name'].values[area4_indexes]

area1_types = IXY['Tectonic type'].values[area1_indexes]
area2_types = IXY['Tectonic type'].values[area2_indexes]
area3_types = IXY['Tectonic type'].values[area3_indexes]
area4_types = IXY['Tectonic type'].values[area4_indexes]
```

```python
def area_value(value_table, value_name, area_names):
    """
    """
    value = []
    tect_type = []
    for name in area_names:
        index = np.where(value_table['name'].values == name)[0][0]
        
        value.append(value_table[value_name].values[index])
        tect_type.append(value_table['tectonic type'].values[index])
        
    return value, tect_type
```

```python
s1, t1 = area_value(entropy_table, 'entropy', area1_names)
s2, t2 = area_value(entropy_table, 'entropy', area2_names)
s3, t3 = area_value(entropy_table, 'entropy', area3_names)
s4, t4 = area_value(entropy_table, 'entropy', area4_names)

s1 = np.array(s1).astype('float')
t1 = np.array(t1)

s2 = np.array(s2).astype('float')
t2 = np.array(t2)

s3 = np.array(s3).astype('float')
t3 = np.array(t3)

s4 = np.array(s4).astype('float')
t4 = np.array(t4)
```

```python
import ternary
```

```python
def norm(array):
    """
    Нормирует количества точек X, I, Y от 0 до 1
    """
    n = len(array)
    array_normed = np.zeros(array.shape)
    for i in range(n):
        array_normed[i] = array[i, :] / np.sum(array[i, :])
        
    return array_normed
```

```python
counts = df[['X_nodes', 'I_nodes', 'Y_nodes']]
counts_normed = norm(counts.values)
counts_normed_I = counts_normed[:, 1]
```

```python
# области тернарной диаграммы
area1_indexes = np.where(counts_normed_I > 0.8)[0]

area2_indexes = np.where((counts_normed_I > 0.3) & (counts_normed_I < 0.6))[0]

area3_indexes = np.where((counts_normed_I > 0.1) & (counts_normed_I <= 0.3))[0]

area4_indexes = np.where(counts_normed_I < 0.1)[0]
```

```python
# визуализация областей

rcParams['figure.figsize'] = 8, 8
figure, tax = ternary.figure(scale=1.0)
figure.set_facecolor('w')
tax.set_background_color('w')


tax.scatter(counts_normed[area1_indexes], color='C0', label='область 1 (I > 0.8)')
tax.scatter(counts_normed[area2_indexes], color='C1', label='область 2 (0.3 < I < 0.6)')
tax.scatter(counts_normed[area3_indexes], color='C2', label='область 3 (0.1 < I < 0.3)')
tax.scatter(counts_normed[area4_indexes], color='C3', label='область 4 (I < 0.1)')
    

tax.boundary(linewidth=1.0)
tax.ticks(axis='lbr', linewidth=1, multiple=0.1, tick_formats="%.1f")
tax.gridlines(multiple=0.2, color="black")
tax.legend(facecolor='white', fontsize=13.5, bbox_to_anchor=(0.3, 0.95))
tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()

tax.right_corner_label("X-Nodes", fontsize=18, offset=-0.00001)
tax.top_corner_label("                        I-Nodes", fontsize=18, offset=0.1)
tax.left_corner_label("           Y-Nodes", fontsize=18, offset=0.35)

tax.show()
rcParams['figure.figsize'] = 13, 6
```
